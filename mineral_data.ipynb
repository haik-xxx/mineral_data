```python
file_path="C:/Users/HP/Documents/School/Year 4/Sem 1/SGL 411 MINERAL EXPLORATION & MINING/Surfer assignment/assignment 1 data in ppm csv.csv"
```

```python
import csv 
titanium=[]
iron=[]
with open(file_path, newline='') as csvfile:
    next(csvfile)
    next(csvfile)
    reader = csv.reader(csvfile)
    for n in reader:
        titanium.append(int(n[3]))
        iron.append(int(n[7]))
```

```python
import numpy as np
titanium_normalized = np.array(titanium) / np.max(titanium) #normalizing data
iron_normalized = np.array(iron) / np.max(iron)
```

```python
import matplotlib.pyplot as plt
plt.scatter(titanium_normalized,iron_normalized)
plt.xlabel("titanium(ppm)")
plt.ylabel("iron(ppm)")
plt.title("Concentration of iron vs titanium")
plt.grid()
plt.show()
```
